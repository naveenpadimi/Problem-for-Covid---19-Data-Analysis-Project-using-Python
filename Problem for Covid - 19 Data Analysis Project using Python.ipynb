```python
import pandas as pd
import seaborn as sns
```

```python
url = "covid-data.csv"
df = pd.read_csv(url)
df
```

```python
# Step 2: High Level Data Understanding
# a. Find no. of rows & columns in the dataset
```

```python
num_rows, num_cols = df.shape
print("Number of rows:", num_rows)
print("Number of columns:", num_cols)
```

```python
# b. Data types of columns
print("Data types of columns:")
print(df.dtypes)
```

```python
# c. Info & describe of data in dataframe

print("Dataframe info:")
print(df.info())
print("Dataframe description:")
print(df.describe())
```

```python
# Step 3: Low Level Data Understanding
```

```python
# a. Find count of unique values in location column
unique_locations = df['location'].nunique()
print("Count of unique values in location column:", unique_locations)
```

```python
# b. Find which continent has maximum frequency using value counts
continent_frequencies = df['continent'].value_counts()
continent_with_max_frequency = continent_frequencies.idxmax()
print("Continent with maximum frequency:", continent_with_max_frequency)
```

```python
# c. Find maximum & mean value in 'total_cases'
max_total_cases = df['total_cases'].max()
mean_total_cases = df['total_cases'].mean()
print("Maximum total cases:", max_total_cases)
print("Mean total cases:", mean_total_cases)
```

```python
# d. Find 25%, 50%, and 75% quartile values in 'total_deaths'
quartiles_total_deaths = df['total_deaths'].quantile([0.25, 0.5, 0.75])
print("25% quartile of total deaths:", quartiles_total_deaths[0.25])
print("50% quartile of total deaths:", quartiles_total_deaths[0.5])
print("75% quartile of total deaths:", quartiles_total_deaths[0.75])
```

```python
# e. Find which continent has maximum 'human_development_index'
continent_with_max_hdi = df.loc[df['human_development_index'].idxmax(), 'continent']
print("Continent with maximum human development index:", continent_with_max_hdi)
```

```python
# f. Find which continent has minimum 'gdp_per_capita'
continent_with_min_gdp = df.loc[df['gdp_per_capita'].idxmin(), 'continent']
print("Continent with minimum GDP per capita:", continent_with_min_gdp)
```

```python
# Step 4: Filter the dataframe with only selected columns
selected_columns = ['continent', 'location', 'date', 'total_cases', 'total_deaths', 'gdp_per_capita', 'human_development_index']
df = df[selected_columns]
df
```

```python
# Step 5: Data Cleaning
```

```python
# a. Remove all duplicate observations
df.drop_duplicates(inplace=True)
```

```python
# b. Find missing values in all columns
missing_values = df.isnull().sum()
print("Missing values in each column:")
print(missing_values)
```

```python
# c. Remove all observations where continent column value is missing
df.dropna(subset=['continent'], inplace=True)
```

```python
# d. Fill all missing values with 0
df.fillna(0, inplace=True)
```

```python
# Step 6: Date time format
```

```python
# a. Convert date column to datetime format using pandas.to_datetime
df['date'] = pd.to_datetime(df['date'])
```

```python
# b. Create a new column 'month' after extracting month data from the date column
df['month'] = df['date'].dt.month
```

```python
# Step 7: Data Aggregation
```

```python
# a. Find max value in all columns using groupby function on 'continent' column
df_groupby = df.groupby('continent').max().reset_index()
```

```python
# Step 8: Feature Engineering
```

```python
# a. Create a new feature 'total_deaths_to_total_cases' by ratio of 'total_deaths' to 'total_cases'
df_groupby['total_deaths_to_total_cases'] = df_groupby['total_deaths'] / df_groupby['total_cases']
```

```python
# Step 9: Data Visualization
```

```python
# a. Perform Univariate analysis on 'gdp_per_capita' column by plotting histogram using seaborn dist plot
sns.displot(df_groupby['gdp_per_capita'])
```

```python
# b. Plot a scatter plot of 'total_cases' vs 'gdp_per_capita'
sns.scatterplot(data=df_groupby, x='total_cases', y='gdp_per_capita')
```

```python
# c. Plot Pairplot on df_groupby dataset
sns.pairplot(df_groupby)
```

```python
# d. Plot a bar plot of 'continent' column with 'total_cases'
sns.catplot(data=df_groupby, x='continent', y='total_cases', kind='bar')
```

```python
# Step 10: Save the df_groupby dataframe to your local drive using pandas.to_csv function
df_groupby.to_csv('df_groupby.csv', index=False)
```
